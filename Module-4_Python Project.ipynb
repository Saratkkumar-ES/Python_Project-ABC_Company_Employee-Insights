```python
# Importing required files 
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading Excel file 
df = pd.read_excel("C:/Users/user/Desktop/ASSIGNMENTS-ENTRI/Python- Project/myexcel.xlsx")
# Display the first few rows
print(df.head())
```

## Preprocessing

```python
# Correct the data in the "height" column by replacing it with random numbers between 150 and 180. 

df["Height"] = np.random.randint(150, 181, size=len(df))

# Display updated data
print(df.head())
```

```python
# View Summary of the Data
print(df.info())
```

```python
print(df.describe()) 
```

```python
# Filling the null values in salary Column
# Calculate the mean salary (ignoring NaN values)
mean_salary = df["Salary"].mean()


# Fill NaN values in the "Salary" column with the mean salary

df["Salary"] = df["Salary"].fillna(mean_salary)
# Print the updated DataFrame
print(df)
```

```python
# Filling the null values in College with Mode.

# Calculate the mode (most frequent value) for the "College" column
mode_college = df["College"].mode()[0]           # mode returns a series, so we take the first value

# Fill NaN values in the College column with the mode
df["College"] = df["College"].fillna(mode_college)
# Print the updated DataFrame
print(df)
```

```python
df.to_csv("updated_file.csv", index=False)
```

## Analysis Tasks


```python
# To count the number of employees in each team
team_distribution = df["Team"].value_counts()


# Calculate the percentage of each team
team_percentage = (team_distribution / len(df)) * 100
# Create a DataFrame to display both counts and percentages
team_summary = pd.DataFrame({"Count": team_distribution, "Percentage": team_percentage})

# Display
print(team_summary)
```

```python
# Visualization 
# Ploting Bar Chart
plt.figure(figsize=(10, 6))

sns.barplot(x=team_summary.index, y=team_summary["Count"], hue=team_summary.index, palette="viridis", legend=False)

# labels and title
plt.xlabel("Team")
plt.ylabel("Number of Employees")
plt.title("Distribution of Employees Across Each Team")
# Rotate x labels for readability
plt.xticks(rotation=90) 

# Show plot
plt.show()
```

```python
# Count the number of employees in each position
employee_position = df["Position"].value_counts()

# Calculate the percentage of employees in each position
employee_percentage = (employee_position / len(df)) * 100

# Create a DataFrame to display both counts and percentages
position_summary = pd.DataFrame({"Count": employee_position, "Percentage": employee_percentage})

# Display
print(position_summary)
```

```python
# Ploting Donut Chart

# Group by Position and count the number of employees
position_summary = df["Position"].value_counts()

# Plot a pie chart
plt.figure(figsize=(8, 8))
plt.pie(position_summary, labels=position_summary.index, autopct='%1.1f%%', 
        colors=plt.cm.tab20.colors[:len(position_summary)], startangle=140, wedgeprops={'width': 0.3})
plt.title("Total Employees by Position")
plt.show()
```

3. Identify the predominant age group among employees

```python
# To identify the predominant age group I define age groups in to 18-25, 26-35, 36-45,..
# Creating age groups
bins = [0, 25, 30, 35, 40, 45] 
labels = ['18-25', '26-30', '31-35', '36-40', '40-45']

# Categorizing the ages
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)

# to view count of employees in each age group
age_group_counts = df['Age Group'].value_counts().sort_index()

print(age_group_counts)

# the predominant age group
predominant_age_group = age_group_count.idxmax()
print(f"The predominant age group is: {predominant_age_group}")
```

```python
# Ploting Pie Chart

# Ploting the distribution using a pie chart
plt.figure(figsize=(8, 6))
plt.pie(age_group_count, labels=age_group_count.index, autopct='%1.1f%%', 
        colors=plt.cm.Paired.colors, startangle=140)
plt.title('Employees by Age Group')
plt.show()

# the predominant age group
predominant_age_group = age_group_count.idxmax()
print(f"The predominant age group is: {predominant_age_group}")
```

```python
# 4 Discover which team and position have the highest salary expenditure
```

```python

# Grouping by Team to find total salary expenditure per team
team_salary = df.groupby("Team")["Salary"].sum().sort_values(ascending=False)


# Displaying the top salary expenditure for Teams
print("Salary Expenditure by Team:")
print(team_salary.reset_index())
```

```python
# Grouping by Position to find total salary expenditure per position
position_salary = df.groupby("Position")["Salary"].sum().sort_values(ascending=False)

# Displaying the top salary expenditure for Positions
print("\nSalary Expenditure by Position:")
print(position_salary.reset_index())
```

```python
# Setting figure size
plt.figure(figsize=(10, 6))

# Bar Chart for Team Salary Expenditure
sns.barplot(x=team_salary.head(10).index, y=team_salary.head(10).values, hue=team_salary.head(10).index, 
            palette="viridis", legend=False)

plt.xticks(rotation=90)
plt.title("Top Teams by Salary Expenditure")
plt.xlabel("Team")
plt.ylabel("Total Salary")

plt.show()
```

```python
# Setting figure size
plt.figure(figsize=(8, 6))

# Pie Chart for Position Salary Expenditure
colors = plt.cm.Paired.colors  # Custom colors
plt.pie(position_salary.head(10).values, labels=position_salary.head(10).index, autopct='%1.1f%%',
        colors=colors, startangle=140)

plt.title("Positions by Salary Expenditure")

plt.show()
```

```python
# 5 Investigate if there's any correlation between age and salary, and represent it visually
```

```python
# Calculating the correlation coefficient
correlation = df["Age"].corr(df["Salary"])
print(f"Correlation between Age and Salary: {correlation:.2f}")
```

```python
# Ploting a scatter plot with a regression line
plt.figure(figsize=(8, 6))
sns.regplot(x=df["Age"], y=df["Salary"], scatter_kws={"alpha": 0.5}, line_kws={"color": "red"})
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Correlation between Age and Salary")
plt.grid(True)
plt.show()
```
